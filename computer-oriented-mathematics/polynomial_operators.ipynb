```python
# Aufgabe 5
# a)
# i)

#p = [a0, a1, ..., ak]

# helper funct to get max length given two lists
def get_max(p,q):
    if len(p) >= len(q): return len(p)
    return len(q)

def poly_add(p,q):
    n = get_max(p,q)
    p1 = p + [0]*(n-len(p))
    q1 = q + [0]*(n-len(q))
    sum = [p1[i] + q1[i] for i in [0..(n-1)]]
    return sum

def poly_sub(p,q):
    n = get_max(p,q)
    p1 = p + [0]*(n-len(p))
    q1 = q + [0]*(n-len(q))
    sub = [p1[i] - q1[i] for i in [0..(n-1)]]
    return sub

p = [3,2]   # 3 + 2x 
q = [2,1]   # 2 + 1x
print(poly_add(p,q)) # expected: [5,3]
print(poly_sub(p,q)) # expected: [1,1]
print(poly_sub((p+[2]),q)) # expected = [1,1,2]



```

```python
# ii)
def poly_mulp(p,q):

    new_grad = len(p) + len(q) - 2
    result = []
    
    for k in range(new_grad +1):
        c = 0
        for i in range(len(p)):
            j = k-i
            if 0<= j < len(q):
                c += p[i]*q[j]
        result.append(c)

    return result



p = [3,2]   # 3 + 2x 
q = [2,1]   # 2 + 1x
print(poly_mulp(p,q)) # expected: [6,7,2]
```

```python
# Helper function: remove trailing zeros
def rem_nulls(p):
    while p and p[-1] == 0:
        p.pop()
    return p

print(rem_nulls([2,0,0,1,0,0,0]))  # Expected: [2,0,0,1]

# Polynomial division (returns quotient and remainder)
def poly_divmod(p, q):
    p = rem_nulls(p[:])  # copy and remove trailing zeros
    q = rem_nulls(q[:])
    
    if len(p) < len(q):
        return ([0], p)  # quotient 0, remainder p
    
    m = p[:]
    n = len(p) - len(q)
    quotient = [0]*(n+1)
    
    while len(m) >= len(q):
        factor = m[-1] / q[-1]  # leading coefficient division
        degree_diff = len(m) - len(q)
        quotient[degree_diff] = factor
        
        # Subtract factor * q * x^degree_diff from m
        for i in range(len(q)):
            m[i + degree_diff] -= factor * q[i]
        
        m = rem_nulls(m)
    
    return (quotient, m)

# Polynomial remainder
def poly_mod(p, m):
    return poly_divmod(p, m)[1]

# Polynomial division quotient only
def poly_div(p, m):
    return poly_divmod(p, m)[0]

# Test
p = [1, 2, 3]  # 1 + 2x + 3x^2
m = [2, 1]     # 2 + x
print("Remainder:", poly_mod(p, m))  # Should print remainder of division
print("Quotient:", poly_div(p, m))   # Should print quotient
```

```python
# ii)
def poly_skal(p,s):
    return [x*s for x in p]

def poly_norm(p):
    #last_koeff = p[-1] 
    return poly_skal(p,1/p[-1])

print(poly_skal([2,3,4],2)) # [4, 6, 8]
print(poly_norm([2,3,4])) # [1/2, 3/4, 1]
```

```python
# iii)
def poly_eval(p,x):
    if(len(p) ==  0): return 0
    return p[0] + x * poly_eval(p[1:],x)
    
p = [2,3,4]
print(poly_eval(p,2)) # 24
    
```

```python
# iv)
def poly_abl(p):
    # p = [7,5,3] :  7 + 5x + 3x^2
    # p' = [5,6]     :  5.1 + 3.2
    result = []
    if len(p) > 1:
        for i in range(1,len(p)):
            result.append(i * p[i]) 
        return result
    return [0]
print(poly_abl([7,5,3])) # [5, 6]
```

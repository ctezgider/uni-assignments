```python
# a) – Fibonacci numbers (recursive and iterative)

import time

# Recursive Fibonacci
def fib_recursive(n):
    if n == 0:
        return 1
    elif n == 1:
        return 1
    else:
        return fib_recursive(n - 1) + fib_recursive(n - 2)

# Iterative Fibonacci
def fib_iterative(n):
    if n == 0 or n == 1:
        return 1
    a = 1  # second-to-last element
    b = 1  # last element

    for i in range(n - 1):
        c = a + b
        a = b
        b = c
    return c

# Measure execution time for recursive version
start = time.time()
k = 5
fib_recursive(2**k)
end = time.time()
print("Time:", end - start)
# For k = 5, should be under 10 seconds

# Test iterative function
print("fib_iterative(5) =", fib_iterative(5))
print("fib_recursive(5) =", fib_recursive(5))
```

```python
# c) – Efficient Fibonacci computation (returns (F_n, F_{n+1}))

def fib_efficient(n):
    if n == 0:
        return (0, 1)
    
    k = n // 2
    (Fk, Fk1) = fib_efficient(k)
    
    F2k = Fk * (2 * Fk1 - Fk)
    F2k1 = Fk1^2 + Fk^2
    
    if n % 2 == 0:
        return (F2k, F2k1)
    else:
        return (F2k1, F2k + F2k1)

# Test
print(fib_efficient(4))  # Expected: (3, 5)

# Compute last 10 digits for n = 2^20 - 1
n = 2^20 - 1
last_ten = fib_efficient(n)[0] % (10^10)
print(last_ten)
```

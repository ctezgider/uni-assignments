```python

# Compute greatest common divisor using Euclidean algorithm
def gcd(a,b):
    #codnifitons and their return values accorting to Euklidischer Algorithmus
    if (a == 0): return b
    if (b == 0): return a
    if (a > b): return gcd(a - b ,b)
    return gcd(a, b - a)

# Test
gcd(1234,5678)
    
```

```python
# Compute all divisors of a number
def divisors(y):
    
    # Initialize a list to save all divisors of y
    result = []
    
    # For loop from 1 to y   
    for i in range(1, y + 1):
        if (y % i == 0):
            result.append(i)
    
    # Return the saved list of divisors of y
    return result
    
# Test
divisors(180)
```

```python
# Check if a number is prime

def is_prime(y):
    # Get all divisors of n
    divs = divisors(y)
    
    # Due to def of Primzahl, exclude 1 
    if y == 1 : return False

    # Remove 1 and given y from list, bcz they are always divisors of y
    divs.remove(1)
    divs.remove(y)
    
    # Check if length of list
    # If zero, it is prime, else not because there are other divisors of y
    if len(divs) == 0: return True
    return False

# Test 
is_prime(157)   # True
```

```python
# Compute prime factors with their powers
def prime_factors(y):
    # List of tuples like (prime factor, its exponent)
    factors = []
    
    n = y
    p = 2
    
    while p * p <= n:  # Only go up to sqrt(n)
        count = 0
        # Count how many times p divides n
        while n % p == 0:
            n //= p  # Divide n by p
            count += 1  # Increase exponent
        if count > 0:  # If p is a prime factor, add to list
            factors.append((p, count))
        p += 1
    
    if n > 1:
        factors.append((n, 1))
    return factors

# Test
[x[0] for x in prime_factors(16)]    # [2]
[x[0] for x in prime_factors(9240)]  # [2, 3, 5, 7]

```

```python
# Find numbers with the maximum number of prime factors in a range
# Initialize list of prime factors for numbers from 0 to 500
list_of_lists = [prime_factors(y) for y in range(0, 501)]

# Get all sizes as a set
sizes = {len(x) for x in list_of_lists}

# Get lists of prime factors with maximum length
max_prime_factors = [
    x for x in list_of_lists
    if len(x) == max(sizes)
]

print("Number : Prime factors")
# Print results
for factor_list in max_prime_factors: 
    num = 1
    # Reconstruct original number from prime factors
    for base, exponent in factor_list:
        num *= base ** exponent
    # Print number and its prime factors
    print(num, ":", [x[0] for x in factor_list])
```

```python

# Main function to calculate all prime numbers from 2 to given n
def sieve_of_eratosthenes(n):

    # Initializing a list from 2 to given n
    list_to_n = [ x for x in (2..n)]
    
    # Remove the nums, which are multiple of x
    for e in list_to_n:
        if is_prime(e):
            list_to_n = remove_mult_nums(e,list_to_n)
        
    return list_to_n



# Helper function to remove multp numbrs
def remove_mult_nums(n,list_nums):
    return [x for x in list_nums if (x == n or x % n !=0)]


# Function to find twin primes up to n
def twin_primes(n):
    # save the prims to given n
    prims_to_n = sieve_of_eratosthenes(n)
    twins = []

    for i in range(0,len(prims_to_n)-1):
        first = prims_to_n[i]
        second = prims_to_n[i+1]
        if (first+2 == second): 
            twins.append((first, second))

    return twins

# Test
twin_primes(10^3)


```

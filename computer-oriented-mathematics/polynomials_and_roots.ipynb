```python
# Finding roots of a cubic polynomial
c = polygen(QQ, 'c')
f = c^3 - c^2 - c - 1

# Compute roots in complex numbers
roots = [x[0] for x in f.roots(ring=CC)]
c1, c2, c3 = roots

# Print roots
print("c1 =", c1)
print("c2 =", c2)
print("c3 =", c3)
```

```python
# Solving a linear system using Gauss elimination

# Construct matrix A and vector b
A = matrix(CC, [
    [1, 1, 1],
    [c1, c2, c3],
    [c1^2, c2^2, c3^2]
])
b = vector(CC, [0, 1, 1])

# Solve the system A * x = b
x = A.solve_right(b)

# Print solution
for i in range(len(x)):
    print(f"b{i+1} =", x[i])
```

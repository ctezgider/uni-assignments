```python
# Implementing function gcd
def gcd(a,b):
    if (a == 0): return b
    if (b == 0): return a
    if (a > b): return gcd(a-b,b)
    return gcd(a,b-a)
```

```python
def dividers(y):
    list = []
    for i in range(1,y+1):
        if (y%i == 0):
            list.append(i)
    return list
```

```python
def Primzahl(y):
    # Initializing a list of Teiler of y
    list = Teiler(y) 
    
    # Due to def of Primzahl, exclude 1 
    if y == 1 : return False

    # remove 1 and given y from list, bcz they are always Teiler of y
    list.remove(1)
    list.remove(y)
    
    # Check if length of list
    # if zero, it is prime, else not because there are other Teiler of y
    if len(list) == 0: return True
    return False

# Test 
print("Is 157 prime?" , Primzahl(157))
print("Is 1023 prime? ", Primzahl(1023))
    
```

```python
def Primteiler(y):
    # init list of Teiler of y
    list_Teiler = Teiler(y)
    
    # for-loop in list_Teiler.copy(), 
    # copy() used, because the orig list is manipulated - remove(i)
    for i in list_Teiler.copy():
        # check if NOT prime number then remove the number from org list
        if not Primzahl(i):
            list_Teiler.remove(i)
    # return manipulated list
    return list_Teiler

# [i for i in list_Teiler if Primzahl(i)]

# Test
print("Primteiler von 9240: " , Primteiler(9240))
```

```python
def Primteiler2(y):
    # list of tupels like (prime divider, its power)
    factors = []
    
    n = y
    p = 2
    
    while p * p <= n: # calculate the sqrt same thing, cz its shorter
        count = 0
        # count to calculate power of prime divider
        while n % p == 0:
            n//=p   # take the rest 
            count +=1 # increase counter
        if count > 0: # if there is a prime divider, add it to the list
            factors.append((p, count))
        p +=1
    
    if n > 1:
        factors.append((n, 1))
    return factors

# TEST
print("Primteiler von 16:", [x[0] for x in Primteiler2(16)])
print("Primteiler von 9240:", [x[0] for x in Primteiler2(9240)])
```

```python
# main funct to calculate all prime numbers from 2 to given n
def Eratosthenes_Sieb(n):

    # initializing a list from 2 to given n
    list_to_n = [ x for x in (2..n)]
    
    # remove the nums, which are multiple of x
    for e in list_to_n:
        if Primzahl(e):
            list_to_n = Remove_Mult_Nums(e,list_to_n)
        
    return list_to_n



# helper function to remove multp numbrs
def Remove_Mult_Nums(n,list_nums):
   
    return [x for x in list_nums if (x == n or x % n !=0)]


# Primzahlzwillinge 
def Primzahlzwillingspaare(n):
    # save the prims to given n
    prims_to_n = Eratosthenes_Sieb(n)
    zwillinge = []

    for i in range(0,len(prims_to_n)-1):
        first = prims_to_n[i]
        second = prims_to_n[i+1]
        if (first+2 == second): 
            zwillinge.append((first, second))

    return zwillinge

# TEST
print(Primzahlzwillingspaare(10^4))


```
